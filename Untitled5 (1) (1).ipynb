<h1 align="center">CALCOLA L'IPOTENUSA</h1> 

### Per prima dichiariamo c1, c2 e ip e tutto reale. Il codice dice al utente di dare valore di c1, dopo legge questo valore e fa stessa cosa con c2, dopo scrive "calco l'ipotenusa" e fa calcolo . Dopo calcolo scrive ad utente che calcolato e da il risultato.

```python
import math
print("dammi un valore per c1")
c1 = float(input())
print("dammi un valore per c2")
c2 = float(input())
print("calco l'ipotenusa")
ip = math.sqrt(c1**2 + c2**2)
print("ipotenusa calcolata")
print(ip)
```

<h1 align="center">UGUALE, MINORE O MAGGIORE</h1> 

### Dichiriamo num1 e num2 interi. Chiediamo ad utente di dare num1 e la programma legge num1 e fa stessa cosa con num2. Fa condizione tra vero e falso.
```python
print("dammi numero 1")
num1 = int(input())
print("dammi numero 2")
num2 = int(input())
if num1 == num2:
    print("due numeri sono uguali")
else:
    if num1 > num2:
        print("num1 Ã¨ masggiore di num2")
    else:
        if num1 < num2:
            print("num1 Ã¨ minore di num2")
```

<h1 align="center">CALCOLO DELL SCONTO</h1> 

### Dichiariamo 3 variabili reale "lordo", "scontano" e "sconto". Chiediamo ad utente di dare importo di lordo, il codice legge lordo  e chiede ad utente di dare importo scontato e legge importo scontato. Dopo programma fa calcolo tra lordo e importo scontato. La programma scrive a utente "la percentuale di sconto Ã¨" e scrive il sconto in percentuale.

```python
print("Dammi l'importo lordo")
lordo = float(input())
print("dammi l'importo scontato")
scontato = float(input())
sconto = 100 - lordo * scontato / 100
print("la percentuale di sconto Ã¨")
print(str(sconto) + "%")
```
